```python
# L=[11, 10, 83, 25, 28]의 원소들을 오른쪽으로 한 칸씩 쉬프트(rotate right)하여 출력하시오. : 리스트 의 확장
while True:
    l = list(map(int, input("list of numbers: ").split()))
    _len = len(l)
    move = int(input("Move: "))
    if move > _len:
        print("move should be smaller than length of the list")
        continue
    
    direction = input("Direction(L/R): ").lower()
    l = l+l
    
    if direction == "l":
        result = l[move:move+_len]
    elif direction == "r":
        result = l[(_len-move):(_len-move)+_len]
    else:
        print("direction should be either L or R")

    print(result)
    
    if input("Continue? Y/N: ").lower() == 'y':
        break
```

```python
class TV:
    def __init__(self, channel, volume, isOn):
        self.channel = channel
        self.volume = volume
        self.isOn = isOn
    
    def turnOn(self):
        self.isOn = True
    
    def turnOff(self):
        self.isOn = False
        
    def setChannel(self, channel):
        self.channel = channel
        
    def setVolume(self, volume):
        self.volume = volume
        
    def status(self):
        print("Channel: ", self.channel)
        print("Volume: ", self.volume)
        print("On/Off: ", self.isOn)
```

```python
tele = TV(7, 6, True)
```

```python
tele.turnOff()
```

```python
tele.status()
```

```python
tele.turnOff()
```

```python
print(tele.status())
```

```python
class Television:
    def __init__(self, channel, volume, isOn):
        self.channel = channel
        self.volume = volume
        self.isOn = isOn

    def turnOn(self):
        self.isOn = True
    
    def turnOff(self):
        self.isOn = False
        
    def setChannel(self, channel):
        self.channel = channel
        
    def setVolume(self, volume):
        self.volume = volume
        
    def status(self):
        print("Channel: ", self.channel)
        print("Volume: ", self.volume)
        print("On/Off: ", self.isOn)
```

```python
tv2 = Television(5, 9, False)
```
