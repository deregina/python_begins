```python
def bubble_sort(list_) :
    length = len(list_)
    for i in range(length) :
        for j in range(length - (i + 1)) :
            if list_[j] > list_[j + 1] : 
                list_[j], list_[j + 1] = list_[j + 1], list_[j]
            else : pass
    return list_
```

```python
import time
import random
```

```python
random_list = []

for i in range(10000) :
    num = random.randrange(1, 10001)
    random_list.append(num)
```

```python
start = time.time()
bubble_sort(random_list)
now = time.time()
print(now - start)
```

```python
def selection_sort(_list) :
    for i in range(len(_list) - 1) :
        select = i
        for j in range(len(_list) - (i + 1)) :
            if _list[select] > _list[i + j + 1] :
                select = i + j + 1
            else : pass
        _list[select], _list[i] = _list[i], _list[select]
```

```python
random_list = []

for i in range(10000) :
    num = random.randrange(1, 10001)
    random_list.append(num)
```

```python
start = time.time()
selection_sort(random_list)
now = time.time()
print(now - start)
```
